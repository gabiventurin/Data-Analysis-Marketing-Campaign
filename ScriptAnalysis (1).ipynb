<h4>1.Data analysis</h4>

```python
#importing the libraries used
import pandas as pd 
import matplotlib.pyplot as plt
#importing the provided dataframe for data analysis
tabela = pd.read_csv("relevanc_data_test.csv")
display(tabela)

print("\n\n\n")
#description of the first data analysis
print("1. The table and graph below are provided to illustrate which products had the most revenue during the \ncampaign period\n") 
print("• It can be seen that product BOLO INGLÊS LARANJA had the highest financial revenue and product BOLO INGLÊS GOTAS DE CHOCOLATE had the second highest revenue")

#grouping products to obtain the total billing
tabela_product_sales = tabela[["product_name", "sales_value"]].groupby(["product_name"]).sum().sort_values(by='sales_value',ascending=False)
print("\nTOTAL SALES TABLE - in ascending order\n")
display(tabela_product_sales)
print("\n\n")
#creating a bar chart to illustrate the table above
#data
product_name = ['BOLO INGLÊS LARANJA', 'BOLOS INGLÊS GOTAS DE CHOCOLATE', 'BOLO INGLÊS FRAPE', 'BOLO INGLÊS CHOCOLATE', 'BOLO REDONDO FUBÁ', 'BOLO REDONDO CHOCOLATE', 'BOLO REDONDO CENOURA', 'BOLO INGLÊS BANANA', 'BOLO REDONDO COCO', 'BOLO INGLÊS COCO','BOLO INGLÊS ABACAXI',]
billing = [111678, 84040, 53338, 46853, 46665, 43116, 38030, 30882, 21272, 19672, 18025]
color = ['#00FF7F']
plt.figure(figsize=(5, 3))
#creating the bar graph
plt.barh(product_name, billing, color=color)
plt.gca().invert_yaxis()
#adding labels
plt.xlabel('billing')
plt.ylabel('product_name')
plt.title('Total revenue per product')
#printing the grafic
plt.show()
print("\n\n\n")



#description of the second data analysis
print("\n\n2. The table below was provided to analyze sales behavior impacted by the campaign, filtering sales by\n group and purchase period")
#grouping sales taking into account the group and purchase period
vendas_por_período= tabela[["product_name", "sales_value", "period", "group"]].groupby(["product_name","period", "group"]).sum().sort_values(by='sales_value',ascending=False).nlargest(5, 'sales_value')

print("\nTOTAL SALES TABLE CONSIDERING THE PERIOD AND GROUP")
print("(in ascending order and considering considering the 5 biggest sales)\n")
display(vendas_por_período)

print("\n\n\n2.1 The graphs below are provided to analyze the percentage of sales impacted by\n the campaign.\n\n")
print(" • Group influenced by the campaign: buyers exposed to the campaign who bought after it ")
print(" • Group not influenced by the campaign: buyers not exposed to the campaign and buyers exposed to \n the campaign who purchased before it\n\n")

#generating a pie chart to illustrate sales in "BOLOS DE LARANJA"
#data
labels = ['Influenced group', 'Uninfluenced group']
quantidades = [70775, 59088]
colors = ['#00FF7F', '#FFDEAD']
#creating a pie chart
plt.figure(figsize=(3, 3))  # Define o tamanho da figura
plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
#adding title
plt.title('Distribution of revenue to "BOLO INGLÊS LARANJA"')
#printing the grafic
plt.axis('equal')  
plt.show()
print("\n\n")


#generating a pie chart to illustrate sales in "BOLOS DE CHOCOLATE"
#data
labels = ['Influenced group', 'Uninfluenced group']
quantidades = [53751, 30369]
colors = ['#00FF7F', '#FFDEAD']
#creating the pie chart
plt.figure(figsize=(3, 3))  # Define o tamanho da figura
plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
#adding title 
plt.title('Distribution of revenue to "BOLO INGLÊS GOTAS DE CHOCOLATE"')
#printing the grafic
plt.axis('equal')  
plt.show()
print("\n\n")




print("\n\n\n\n\n3.The table below is provided to illustrate purchases of key products before\n and after the campaign")
print("•Sales increasing for BOLO INGLÊS LARANJA : 1,09%")
print("•Sales increasing for BOLO INGLÊS GOTAS CHOCOLATE: 1,17%")
print("\n\n")
#Creating a table to check sales growth of "BOLO INGLÊS LARANJA" and "BOLO INGLÊS GOSTAS DE CHOCOLATE"
por_data_e_periodo = tabela[["product_name", "sales_value", "period"]].groupby(["product_name", "period"]).sum().sort_values(by='sales_value',ascending=False).nlargest(10, 'sales_value')
display(por_data_e_periodo)
print("\n\n\n")


#init of the grafics to analise the genders into the campaign 
#inserting the pizza grafic of the sales considering the gender for "BOLO INGLÊS DE LARANJA"
# Data
print("\n\n\n 4. The graphs below were provided to analyze the percentage of men and women influenced by the campaign")
labels = ['Female Costumers', 'Male Costumers']
quantidades = [42202, 28573]
colors = [ 'pink', 'blue']

# Creating the pizza grafic
plt.figure(figsize=(3, 3))  # Define o tamanho da figura
plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Adding the title 
plt.title('Distribution of revenue by Gender - BOLO INGLÊS DE LARANJA')

# printing the grafic
plt.axis('equal')  
plt.show()

#inserting the pizza grafic of the sales considering the gender for "BOLO INGLÊS GOTAS CHOCOLATE"
# Data
labels = ['Female Costumers', 'Male Costumers']
quantidades = [30822, 22848]
colors = [ 'pink', 'blue']

# Creating the pizza grafic
plt.figure(figsize=(3, 3)) # Define o tamanho da figura
plt.pie(quantidades, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Adding the title 
plt.title('Distribution of revenue by Gender - BOLO INGLÊS GOTAS DE CHOCOLATE')

# printing the grafic
plt.axis('equal')  
plt.show()
print("\n\n")
print("4.1 The table below is provided to show revenue before and after the campaign by genre - \nassisting in calculating growth rate")
tabela_com_genero2 = tabela[["sex","period","sales_value","product_name"]].groupby(["sex","period","product_name"]).sum().sort_values(by='sales_value',ascending=False).nlargest(15, 'sales_value')
display(tabela_com_genero2)
print("•Sales increasing for BOLOS DE LARANJA:\n Female: 11%\n Male: 4%\n")
print("•Sales increasing for BOLOS DE CHOCOLATE:\n Female: 23%%\n Male: 9%")
print("\n\n\n")
```
